```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
dates = pd.date_range(start="2020-01-01", periods = 10, freq = 'D')

data = {
    'Date': dates,
    'New York' : np.random.normal(loc=5, scale=10, size=len(dates)),
    'Los Angeles' : np.random.normal(loc=15, scale=8, size=len(dates)),
    'Chicago' : np.random.normal(loc=3, scale=12, size=len(dates)),
    'Houston' : np.random.normal(loc=18, scale=6, size=len(dates)),
    'Phoenix' : np.random.normal(loc=20, scale=5, size=len(dates)),
}

df = pd.DataFrame(data)

print(df)

plt.figure(figsize=(10,6))

for city in cities:
    plt.plot(df['Date'], df[city], label = city)

plt.title('Температура в разных городах в январе 2024 года')
plt.xlabel('Дата')
plt.ylabel('Температура')
plt.xticks(rotation = 45)
plt.grid(True)

plt.legend(title = 'Города')

plt.tight_layout()
plt.show()
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2020]
population = [3.03, 3.68, 4.44, 5.32, 6.12, 6.92, 7.79]
co2_levels = [313, 350, 387, 402, 417, 436, 419]

df2 = pd.DataFrame({
    'years': years,
    'population': population,
    'co2_levels': co2_levels
})

fig, ax1  = plt.subplots(figsize = (10, 6))

ax1.set_xlabel('Год')
ax1.set_ylabel('Мировое население (млрд)', color = 'tab:blue')

ax2 = ax1.twinx()

ax2.set_ylabel('Уровеень CO2', color = 'tab:red')

ax1.plot(df2['years'], df2['population'], color= 'tab:blue', label = 'Мировое население',)
ax2.plot(df2['years'], df2['co2_levels'], color= 'tab:red', marker = 's', label = 'CO2', linestyle = '--')

plt.title('Рост населения и уровень CO2 в 1960 - 2020 годах')

ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper right')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'GOP': [20000, 25000, 30000, 15000, 22000, 28000, 27000, 34000, 15000, 10000],
    'education': [80, 85, 90, 70, 75, 88, 82, 92, 95, 60],
    'healthcare': [75, 80,  85, 65, 70, 90, 88, 93, 60, 58],
    'happiness': [7, 7.5, 8, 6.5, 6.8, 8.2, 7.9, 8.6, 5.9, 6.2],
    'unemployment': [5.4, 6.1, 3.9, 11.5, 7.3, 4.2, 5.1, 3.0, 12.2, 10.5]
}

countries = ['USA', 'Canada', 'Germany', 'Brazil', 'India', 'Australia', 'France', 'Sweden', 'Mexico', 'Sourth Africa']
 
df3 = pd.DataFrame(data, index=countries)

corr = df3.corr()

plt.figure(figsize=(10,8))

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, annot_kws= {'size': 12}, cbar_kws={'shrink': 0.8}, vmin = -1, vmax = 1, square= True)

plt.title('Корреляция между экономическим и социальными показетялми стран')

plt.tight_layout()
plt.show()
```

```python
import numpy as np


np.random.seed(42)
countries = [f"Country_{i}" for i in range(1, 16)]

years = np.arange(2010, 2020)
data = {
    'Country': np.random.choice(countries, size=500),
    'Year': np.random.choice(years, size=500),
    'GOP': np.random.uniform(1e4, 1e5, size=500),
    'Poplation': np.random.uniform(1e6, 1e7, size=500),
    'Unemployment': np.random.uniform(3, 15, size=500),
    'CO2_Emissions': np.random.uniform(1e3, 1e6, size=500)
}

df = pd.DataFrame(data)

gdp_per_year = df.groupby('Year')['GOP'].mean()
pop_per_country = df.groupby('Country')['Poplation'].sum()
unemployemnt_per_year = df.groupby('Year')['Unemployment'].mean()

fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, height_ratios=[2, 1], width_ratios=[3, 2])

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# 1. Линейный график: Средний ВВП по годам
ax1.plot(gdp_per_year.index, gdp_per_year.values, marker='o', linestyle='-', label='Средний ВВП')
ax1.set_title('Средний ВВП по годам')
ax1.set_xlabel('Год')
ax1.set_ylabel('Средний ВВП')
ax1.grid(True)
ax1.legend()

# 2. Гистограмма. Распределение населения по странам
ax2.barh(pop_per_country.index, pop_per_country.values, color = 'skyblue')
ax2.set_title('Общее население по годам')
ax2.set_xlabel('Население')
ax2.set_ylabel('Страны')
ax2.tick_params(axis='x', rotation = 45)

# 3. Пузырьковая диаграмма. Взаимосвязь ВВП и выборосов CO2

bubble_size = df['Poplation'] / 1e6

scatter = ax3.scatter(df['GOP'], df['CO2_Emissions'], s=bubble_size, alpha=0.6, c = df['Unemployment'], cmap='viridis')
ax4.set_title('Тепловая карта безработицы')
ax4.set_xlabel("ВВП")
ax4.set_ylabel('Выбросы CO2')
cbar = plt.colorbar(scatter, ax = ax3)
cbar.set_label('Уровень безработицы {%}')
# 4. Тепловая карта: Средняя безработица по годами и странам
heatmap_data = df.pivot_table(index='Country', columns='Year', values='Unemployment', aggfunc = 'mean')
sns.heatmap(heatmap_data, ax=ax4, cmap='coolwarm', annot=True, fmt = ".1f", cbar_kws={'label': "Берзаботица {%}"})
ax4.set_title('Тепловая карта безработицы')
ax4.set_xlabel("Год")
ax4.set_ylabel('Страна')

plt.tight_layout()
plt.show()


```
